```python
ipl['COST IN ₹ (CR.)'] = ipl['COST IN ₹ (CR.)'].fillna(0)
ipl['Cost IN $ (000)'] = ipl['Cost IN $ (000)'].fillna(0)
```

```python
ipl['2021 Squad'] = ipl['2021 Squad'].fillna('Not Paticipated')
```

```python
#REmove all null values and replace with 0 and Not Paticipated
ipl.isnull().sum()
```

```python
teams = ipl[ipl['COST IN ₹ (CR.)']>0]['Team'].unique()
teams
```

```python
ipl['status'] = ipl['Team'].replace(teams,'sold')
```

```python
#  Participation based on the Role(Batsman, Bowlers, Allrounders and WK)

types = ipl['TYPE'].value_counts()
types.reset_index()

```

```python
plt.pie(types.values, labels=types.index,labeldistance=1.2,autopct='%1.2f%%')
plt.title('Role of Players Participated', fontsize = 15)
plt.plot()
```

```python
# Players Sold and unsold using a bar graph
plt.figure(figsize=(5,5))
fig = sns.countplot(ipl['status'],palette=['Blue','Yellow'])
plt.xlabel('Sold or Unsold')
plt.ylabel('Number of Players')
plt.title('Sold vs Unsold', fontsize=15)
plt.plot()
```

```python
ipl.groupby('status')['Player'].count()
```

```python
#  How many Players Brought by Each Team

plt.figure(figsize=(20,10))
fig=sns.countplot(ipl[ipl['Team']!='Unsold']['Team'])
plt.xlable('Team Name')
plt.ylable('Number of players')
plt.title('Players Brought by each Team', fontsize=12)
plt.xticks(rotation=70)
plt.plot()
```

```python
ipl['retention'] = ipl['Base Price']
```

```python
ipl['retention'].replace(['2 Cr', '40 Lakh', '20 Lakh', '1 Cr', '75 Lakh','50 Lakh', '30 Lakh','1.5 Cr'],'From Auction', inplace = True)
ipl['retention']
```

```python
# Treating Base Price

ipl['base_price_unit'] = ipl['Base Price'].apply(lambda x: str(x).split(' ')[-1])
ipl['base_price'] = ipl['Base Price'].apply(lambda x: str(x).split(' ')[0])
```

```python
ipl['base_price'].replace('retention',0,inplace=True)
```

```python
plt.figure(figsize=(20,10))
fig=sns.countplot(ipl[ipl['Team']!='Unsold']['Team'],hue=ipl['TYPE'])
plt.xlable('Player in Each Team')
plt.ylable('Team Name')
plt.title('Number of Player')
plt.xticks(rotation=70)
```

```python
# Highest amount spent on a single player by each team?

ipl[ipl['retention']=='From Auction'].groupby(['Team'])['COST IN ₹ (CR.)'].max()[:-1].sort_values(ascending = False)
```

```python
#  Player Retained at max price?

ipl[ipl['retention']=='Retained'].sort_values(by = 'COST IN ₹ (CR.)', ascending = False).head(1)
```

```python
ipl = ipl.rename(columns={'TYPE':'Type','COST IN ₹ (CR.)':'Sold_for_lakh','Cost IN $ (000)':'Cost_in_dollars','2021 Squad':'Prev_team','Team':'Curr_team'})
```

```python
#  List of players who Played IPL 2021 but went unsold in IPL 2022

unsold_stars = ipl[(ipl.Prev_team != 'Not Participated in IPL 2021') & (ipl.Curr_team == 'Unsold')][['Player','Prev_team']]
unsold_stars
```

```python
# unsold player by each Team

unsold_stars = unsold_stars.groupby('Prev_team')['Player'].count()
unsold_stars
```
